```python
import numpy as np

N = 16
M = 15

array = np.random.rand(M, N)
print(array)
```

```python
def find_nearest_value(z, A):
    i, j = np.unravel_index(np.abs(A - z).argmin(), A.shape)
    return A[i, j]
z = 0.2
result = find_nearest_value(z, array)
print("Знайдене найближче значення:")
print(result)
```

```python
mean_value = np.mean(array)
median_value = np.median(array)
print("Середнє значення:", mean_value)
print("Медіана:", median_value)
```

```python
import matplotlib.pyplot as plt

plt.imshow(array, cmap="viridis")
plt.show()
```

```python
indices_to_null = np.random.choice(np.arange(M), size=10), np.random.choice(np.arange(N), size=10)
array[indices_to_null] = 0
print("Масив після занулення деяких значень:")
print(array)

mean_array = np.mean(array)
array[indices_to_null] = mean_array
print("Масив після відновлення занулених значень:")
print(array)
```

```python
normalized_array = (array - array.min()) / (array.max() - array.min())
print("Нормалізований масив:")
print(normalized_array)
```

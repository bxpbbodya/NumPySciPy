```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg')  
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import numpy as np

min_value = 1
max_value = 66

image_contrast = np.copy(image)
image_contrast = np.clip(image_contrast, min_value, max_value)  
plt.imshow(image_contrast)
plt.axis('off')
plt.show()
```

```python
r, g, b = cv2.split(image)

plt.figure(figsize=(10, 4))
plt.subplot(131)
plt.imshow(r, cmap="Reds")
plt.axis('off')
plt.title("Червоний (R)")
plt.subplot(132)
plt.imshow(g, cmap="Greens")
plt.axis('off')
plt.title("Зелений (G)")
plt.subplot(133)
plt.imshow(b, cmap="Blues")
plt.axis('off')
plt.title("Синій (B)")
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import gaussian_filter

# Розмиття зображення
blurred_image = gaussian_filter(image, sigma=3)  # Значення sigma може бути змінене
plt.imshow(blurred_image)
plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import sobel

# Edge detection
edge_image = sobel(gray_image)
plt.imshow(edge_image, cmap='gray')
plt.axis('off')
plt.show()
```
